```python
from spleeter.separator import Separator
import pandas as pd
import numpy as np
import os, sys
```

```python
cur_dir = "/content/drive/MyDrive/ColabNotebooks/nlp_conf/" ### 수정 ###
cur_dir = cur_dir + "/노래/"
```

```python
file_list = os.listdir(cur_dir)
mp3_file_list = [file for file in file_list if file.endswith('.mp3')]
mp3_file_list
```

```python
separator = Separator('spleeter:2stems')
```

```python
for file in mp3_file_list:
  file_name = file[:-4]
  separator.separate_to_file(f'{cur_dir}{file_name}.mp3', f'{cur_dir}/mr_remove', codec='mp3', filename_format='{filename}_{instrument}.{codec}')
  print(f"{file} --- done")
```

```python
out_dir = cur_dir + "mr_remove/"
```

```python
out_file_list = os.listdir(out_dir)
out_file_list
```

```python
# MR(accompaniment)파일 삭제
for file in out_file_list:
  if file[-5] == "t":
    os.remove(out_dir+file)
  elif file[-5] == "s":
    rename_file = file[:-11]+".mp3"
    os.rename(out_dir+file, out_dir+rename_file)
  else:
    print(file)
    break
```
